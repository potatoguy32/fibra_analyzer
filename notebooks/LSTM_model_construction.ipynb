```python
# out of the box python modules
import sqlite3
import os
import pathlib

# Data wrangling stuff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from sklearn.preprocessing import MinMaxScaler

# Evaluation metrics
from sklearn.metrics import mean_absolute_error, root_mean_squared_error, max_error

# Neural network stuff
import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

# Internal modules
from src.models.make_model import MyModelStruct


rc('mathtext', default='regular')

%load_ext dotenv
%dotenv

DATABASE_PATH = os.environ['DATABASE_PATH']
MODEL_STORE_PATH = os.environ['MODEL_STORE']
```

```python
with sqlite3.connect(DATABASE_PATH) as conn:
    cur = conn.cursor()
    res = cur.execute("SELECT * FROM market_data where date >= '2020-09-01' order by ticker, date;")
    entries = res.fetchall()

full_dataset = pd.DataFrame(entries, columns=['ticker', 'date', 'open', 'high', 'low', 'close', 'volume', 'dividends', 'stock_splits'])
full_dataset['date'] = pd.to_datetime(full_dataset['date'])
full_dataset['avg_price'] = full_dataset[['open', 'high', 'low', 'close']].mean(axis=1)
full_dataset['paid_dividends'] = full_dataset['dividends'].replace(0, np.nan)
full_dataset['period'] = full_dataset.date.dt.to_period('Q')
full_dataset.set_index('date', inplace=True)
```

```python
for ticker in full_dataset.ticker.unique():
    ticker_name = ticker.split('.')[0]
    sample = full_dataset.query(f"ticker == '{ticker}'")
    dataset = sample.filter(['avg_price'])

    model_struct = MyModelStruct(ticker=ticker_name, model_type='ts', dataset=dataset, scaler=MinMaxScaler(feature_range=(0, 1)))
    model_struct.set_train_test_data(train_size=1, offset=60)

    model_struct.set_model()
    model_struct.train_model(loss='mean_squared_error', epochs=2)
    # model_struct.load_model()
    # model_struct.load_scaler()

    last_observed_date = dataset.index[-1]
    for _ in range(15):
        next = []
        next.append(model_struct._scaler.transform(dataset.values[-60:], ))
        next = np.array(next)
        next = np.reshape(next, (next.shape[0], next.shape[1], 1))

        next_prediction = model_struct._model.predict(next, verbose=False)
        next_prediction = model_struct._scaler.inverse_transform(next_prediction)

        dataset.loc[dataset.index[-1] + pd.Timedelta(days=1)] = next_prediction.item()

    dataset.loc[:last_observed_date + pd.Timedelta(days=1), 'value_type'] = 'Observed'
    dataset.loc[last_observed_date + pd.Timedelta(days=1):, 'value_type'] = 'Predicted'

    model_struct.save_model()
    model_struct.save_scaler()
    #visualize the data
    plt.figure(figsize=(18,6))
    plt.title(f'{ticker_name} model')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Price', fontsize=18)
    plt.plot(dataset.query("value_type == 'Observed'").filter(['avg_price']).iloc[-80:])
    plt.plot(dataset.query("value_type == 'Predicted'").filter(['avg_price']))
    plt.legend(['Values', 'Predictions'], loc='lower right')
    plt.show()
```

```python
performance_metrics = {}
for ticker in full_dataset.ticker.unique():
    current_data = full_dataset.query(f"ticker == '{ticker}'")
    current_dataset = current_data.filter(['avg_price'])
    ticker_name = ticker.split('.')[0]
    
    current_model_struct = MyModelStruct(ticker=ticker_name, model_type='ts', scaler=MinMaxScaler(feature_range=(0, 1)), dataset=current_dataset)
    current_model_struct.set_train_test_data(offset=60)
    
    # model = Sequential()

    # model.add(LSTM(units = 50, return_sequences = True, input_shape = (current_model_struct._x_train.shape[1], 1)))
    # model.add(Dropout(0.25))

    # model.add(LSTM(units = 50, return_sequences = True))
    # model.add(Dropout(0.25))

    # model.add(LSTM(units = 50, return_sequences = True))
    # model.add(Dropout(0.25))

    # model.add(LSTM(units = 50))
    # model.add(Dropout(0.25))

    # model.add(Dense(units = 1))
    
    current_model_struct.set_model()
    current_model_struct.train_model(loss='mean_squared_error', epochs=1)

    predictions = current_model_struct._model.predict(current_model_struct._x_test)
    predictions = current_model_struct._scaler.inverse_transform(predictions)
    
    rmse = root_mean_squared_error(current_model_struct._y_test, predictions)
    mae = mean_absolute_error(current_model_struct._y_test, predictions)
    error_max = max_error(current_model_struct._y_test, predictions)
    
    performance_metrics[ticker_name] = {
        'RMSE': rmse,
        'MAE': mae,
        'max_error': error_max
        }
    
    current_model_struct.save_scaler()
    current_model_struct.save_model()
    
    train = current_dataset[:current_model_struct._training_data_len].copy()
    valid = current_dataset[current_model_struct._training_data_len:].copy()
    valid['Predictions'] = predictions
    
    plt.figure(figsize=(18,6))
    plt.title(f'{ticker} model')
    plt.xlabel('Date', fontsize=18)
    plt.ylabel('Price', fontsize=18)
    plt.plot(train['avg_price'])
    plt.plot(valid[['avg_price','Predictions']])
    plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
    plt.show()
```

```python
performance_metrics_df = pd.DataFrame(performance_metrics.values(), index=performance_metrics.keys())
```
